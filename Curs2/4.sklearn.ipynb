```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.datasets import make_regression
```

```python
make_regression(n_samples=2, n_features=1)
```

```python
make_regression(n_samples=3, n_features=1)
```

```python
make_regression(n_samples=3, n_features=2)
```

```python
x, y = make_regression(n_samples=10, n_features=1)
plt.scatter(x, y);
```

```python
model = LinearRegression()
model
```

```python
model.fit(x, y)
```

```python
plt.scatter(x, y);
plt.plot(x, model.coef_ * x + model.intercept_)
```

```python
x, y = make_regression(n_samples=100, n_features=1, noise=10)
model = LinearRegression()
model.fit(x, y)
plt.scatter(x, y);
plt.plot(x, model.coef_ * x + model.intercept_)
```

```python
model.predict([[1]])
```

```python
knn_model = KNeighborsRegressor()
knn_model
```

```python
knn_model.fit(x, y)
```

```python
knn_model.predict([[1]])
```
