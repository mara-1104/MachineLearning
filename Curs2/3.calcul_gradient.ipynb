```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams["figure.figsize"] = [3.3, 3.3]
```

```python
apartamente = [(37, 65), (55, 80), (80, 120), (60, 95), (92, 205), (41, 75), (170, 400), (145, 367), (82, 205), (72, 198)]

apartamente = np.array(apartamente)

apartamente = apartamente[apartamente[:,0].argsort()]
suprafete, preturi = apartamente[:, 0], apartamente[:, 1]
```

```python
def lin_plot(x, y, a, b):
    plt.scatter(x, y);
    f = lambda x: a * x + b
    plt.plot(x, list(map(f, x)));
```

```python
a=0
b=0
lin_plot(suprafete, preturi, a, b)
```

```python
preturi_calculate = a * suprafete + b

suma_reziduala = 1 / (2 * preturi.size) * np.sqrt(((preturi - preturi_calculate) ** 2).sum())
suma_reziduala
```

```python
derivata_a = 2 * np.sum((preturi_calculate - preturi) * suprafete)
derivata_b = 2 * np.sum((preturi_calculate - preturi))

"de_a", derivata_a, "de_b", derivata_b
```

```python
LEARNING_RATE = 0.001
```

```python
noul_a = a - derivata_a * LEARNING_RATE
noul_b = b - derivata_b * LEARNING_RATE
```

```python
lin_plot(suprafete, preturi, noul_a, noul_b)
```

```python
a=0
b=0

cost = []

def calculeaza_gradient():
    global a, b, cost
    LEARNING_RATE = 0.00001
    preturi_calculate = a * suprafete + b

    suma_reziduala = (1 / (2 * preturi.size)) * np.sqrt(((preturi - preturi_calculate) ** 2).sum())
    cost.append(suma_reziduala)
    print("noua suma reziduala:", suma_reziduala)

    derivata_a = (1/preturi.size) * np.sum((preturi_calculate - preturi) * suprafete)
    derivata_b = (1/preturi.size) * np.sum((preturi_calculate - preturi))

    a = a - derivata_a * LEARNING_RATE
    b = b - derivata_b * LEARNING_RATE
    
    lin_plot(suprafete, preturi, a, b)
```

```python
lin_plot(suprafete, preturi, a, b)
```

```python
calculeaza_gradient()
```
